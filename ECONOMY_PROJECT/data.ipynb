```python
import os
import pandas as pd
import numpy as np

DATA_DIR = "data"
FILES = {
    "usdkrw": "DEXKOUS.csv",
    "dgs2": "DGS2.csv",
    "dgs10": "DGS10.csv",
    "dfii10": "DFII10.csv",
    "t10yie": "T10YIE.csv",
    "vix": "VIXCLS.csv",
    "stlfsi": "STLFSI4.csv",
    "dxy": "DTWEXBGS.csv",
}

print("현재 폴더:", os.getcwd())
print("data 폴더 파일:", os.listdir(DATA_DIR))

for k, v in FILES.items():
    path = os.path.join(DATA_DIR, v)
    print(k, "->", path, "| exists:", os.path.exists(path))
```

```python
path = os.path.join(DATA_DIR, FILES["usdkrw"])

df = pd.read_csv(path)

df.columns = ["date", "usdkrw"]

df["date"] = pd.to_datetime(df["date"])

df = df.set_index("date")

df.head()

# df.plot(figsize=(10,4))

# list(df.columns)

# df.info()
```

```python
dfs = {}

for key, file in FILES.items():
    
    path = os.path.join(DATA_DIR, file)
    
    df = pd.read_csv(path)
    
    # 컬럼 이름 정리
    df.columns = ["date", key]
    
    # 날짜 변환
    df["date"] = pd.to_datetime(df["date"])
    
    # index 설정
    df = df.set_index("date")
    
    dfs[key] = df
    
    print(f"\n===== {key} =====")
    print(df.head())
```
