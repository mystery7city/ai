```python
import os
import pandas as pd
import numpy as np

DATA_DIR = "data"
FILES = {
    "usdkrw": "DEXKOUS.csv",
    "dgs2": "DGS2.csv",
    "dgs10": "DGS10.csv",
    "dfii10": "DFII10.csv",
    "t10yie": "T10YIE.csv",
    "vix": "VIXCLS.csv",
    "stlfsi": "STLFSI4.csv",
    "dxy": "DTWEXBGS.csv",
}

print("현재 폴더:", os.getcwd())
print("data 폴더 파일:", os.listdir(DATA_DIR))

for k, v in FILES.items():
    path = os.path.join(DATA_DIR, v)
    print(k, "->", path, "| exists:", os.path.exists(path))
```

```python
dfs = {}

for key, file in FILES.items():
    
    path = os.path.join(DATA_DIR, file)
    
    df = pd.read_csv(path)
    
    # 컬럼 이름 정리
    df.columns = ["date", key]
    
    # 날짜 변환
    df["date"] = pd.to_datetime(df["date"])
    
    # index 설정
    df = df.set_index("date")
    
    dfs[key] = df
    
    print(f"\n===== {key} =====")
    print(df.head())
```

```python
display(dfs["usdkrw"].head(10))
display(dfs["usdkrw"].tail(10))
dfs["usdkrw"].info()
dfs["usdkrw"].describe()
dfs["usdkrw"].isnull().sum()
```

```python
dfs = {}

for key, file in FILES.items():
    path = os.path.join(DATA_DIR, file)
    df = pd.read_csv(path)

    df.columns = ["date", key]
    df["date"] = pd.to_datetime(df["date"])
    df = df.set_index("date").sort_index()

    dfs[key] = df
```

```python
df_all = pd.concat(dfs.values(), axis=1)
df_all.head()
```

```python
summary = []

for key, df in dfs.items():
    summary.append({
        "name": key,
        "start": df.index.min(),
        "end": df.index.max(),
        "rows": len(df)
    })

pd.DataFrame(summary)
```

```python
df_all = pd.concat(dfs.values(), axis=1)
df_all = df_all.sort_index()

df_clean = df_all.dropna()

df_clean.head()
```
