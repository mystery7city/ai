## 1.3 그래프 영역 나누기
### 1) subplot()함수로 서브플롯 추가

```python
import numpy as np
x = np.arange(0,10,0.1) # np.linspace(0,9.9,100)
x.shape
```

```python
plt.figure(figsize=(5,5))
plt.subplot(2, 1, 1) # 2행 1열로 subplot에서 1번째 subplot은 다음과 같이 그린다
plt.plot(x, np.sin(x))
plt.subplot(2,2,3) # 2행2열 subplot중 3번쨰 subplot
plt.plot(x, np.cos(x))
plt.subplot(2,2,4) # 2행 2열 subplot wnd 4번째 subplot
plt.scatter(x, np.sin(x) * np.cos(x))
plt.show()
```

```python
plt.subplot(2, 2, 1); plt.plot(x, np.sin(x))
plt.subplot(2, 2, 2); plt.plot(x, np.cos(x))
plt.subplot(2, 2, 3); plt.plot(x, np.sin(x))
plt.subplot(2, 2, 4); plt.plot(x, np.sin(x) * np.cos(x))
plt.show()
```

### 2) subplots()함수로 서브플롯 추가
- plot함수에 생상이나 스타일 조정(교안 14page)
- https://wikidocs.net/92085

```python
# 2행 2열 서브플롯
fig, axes = plt.subplots(nrows = 2, ncols = 2, figsize=(7,4))
axes[0,0].plot(x, np.sin(x), c='r', linewidth=2)
axes[0,1].plot(x, np.cos(x), 'c--') #linestyle = '--'
axes[1,0].plot(x, np.tanh(x) , '#00AA00')
axes[1,0].plot([0,1,2,3],[-1,-4,5,0.5])
axes[1,1].plot(x, np.sin(x) * np.cos(x), 'r-.') # k:black, marker='o',linestyle=':''
plt.show()



```

```python
# 2행2열 서브플롯(for)
func = [np.sin, np.cos, np.tanh, lambda x:np.sin(x)*np.cos(x)]
fmt  = ['r', 'c--', '#00AA00', 'r-.']
fig, axes = plt.subplots(2,2,figsize=(7,4))
list(enumerate(axes.flat))
for i, ax in enumerate(axes.flat):
    ax.plot(x, func[i](x), fmt[i])
```

```python
# 4행1열 서브플롯(for)
func = [np.sin, np.cos, np.tanh, lambda x:np.sin(x)*np.cos(x)]
fmt  = ['r', 'c--', '#00AA00', 'r-.']
fig, axes = plt.subplots(4,1,figsize=(7,4))


for i, ax in enumerate(axes):
    ax.plot(x, func[i](x), fmt[i])
```
