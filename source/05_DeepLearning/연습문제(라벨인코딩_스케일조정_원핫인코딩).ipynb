# 2. 지역명의 라벨인코딩
-지역명2

```python
X = df[['지역명']].values
```

```python
from sklearn.preprocessing import LabelEncoder


X_flat = X.ravel()  # 또는 X.flatten()

le = LabelEncoder()
labeling_data = le.fit_transform(X_flat)

print('라벨인코딩된 데이터 :', labeling_data)
```

```python
df['지역명2'] = le.fit_transform(df['지역명'])
```

# 3. MinMaxScale조정


```python
X = df[['지역명2', '연도','월']].values # numpy배열로 추출한 독립변수
y = df[['평당분양가격']].values
X.shape, y.shape
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, MinMaxScaler

scaler = MinMaxScaler()


df[['지역명2m', '연도m', '월m']] = scaler.fit_transform(df[['지역명2', '연도', '월']])


print(df)
```

```python
# pip install tabulate
from tabulate import tabulate

print(tabulate(df.head(10), headers='keys', tablefmt='grid', showindex=False))
```

4. standardScale 조정

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler


scaler = StandardScaler()


df[['지역명2s', '연도s', '월s']] = scaler.fit_transform(df[['지역명2', '연도', '월']])


print(df)
```
