```python
## 복합체인 : 나라명 -> 그 나라에서 제일 유명한 음식(food_chain)
##                      음식 -> 레시피(recipe_chain)
```

```python
from langchain_core.output_parsers import StrOutputParser
from langchain_ollama import ChatOllama
from langchain_core.prompts import PromptTemplate

llm = ChatOllama(model="llama3.2",
                temperature=0) # 일관된 답변(보수적인 답변)


# prompt_template = PromptTemplate(
#     template = "What is the capital of {country}. Retrun the name of the city only.",
#     input_variables = ["country"]
# )
# output_parser = StrOutputParser() # AIMessage()를 Str변환
# output_parser.invoke(llm.invoke(prompt_template.invoke({"country":"Korea"})))
```

```python
from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import PromptTemplate

food_template = PromptTemplate(
    template = "What is the most famous food in the {country}. Retrun the name of the food only.",
    input_variables = ["country"]
)

country = input("나라를 입력하세요: ")
prompt = food_template.invoke({"country": country})

ai_message = llm.invoke(prompt)

output_parser = StrOutputParser()
result = output_parser.invoke(ai_message)
result
```

```python
from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import PromptTemplate

recipe_template = PromptTemplate(
    template = "describe the recipe of the {food}.Explain in 4 lines or less",
    input_variables = ["food"]
)

food = input("음식 이름을 입력하세요: ")
prompt = recipe_template.invoke({"food": food})

ai_message = llm.invoke(prompt)


output_parser = StrOutputParser()
result = output_parser.invoke(ai_message)
result
```

```python
food_chain = food_template | llm | output_parser

country = input("나라를 입력하세요: ")
food_chain.invoke({"country":country})
```

```python
recipe_chain = recipe_template | llm | output_parser

food = input("음식을 입력하세요:")
recipe_chain.invoke({"food":food})
```

```python
final_chain = food_chain | recipe_chain
country = input("나라를 입력하세요: ")
final_chain.invoke({"country":country})
```
