```python
data["dong_id"] = data["법정동명"].astype("category").cat.codes
```

```python
data[["법정동명", "dong_id"]].drop_duplicates().head(10)
```

```python
data.head()
```

```python
cols = list(data.columns)

# '법정동명' 위치 찾기
idx = cols.index("법정동명")

# 'dong_id'를 꺼내서, 바로 오른쪽에 삽입
cols.insert(idx + 1, cols.pop(cols.index("dong_id")))

# 컬럼 순서 적용
data = data[cols]
```

```python
data.head()
```

```python
data[["법정동명", "dong_id"]].head()
```

```python
data.head()
```

```python
data["date"] = pd.to_datetime(data["date"])
```

```python
data["year"] = data["date"].dt.year
data["month"] = data["date"].dt.month
data["day"] = data["date"].dt.day
# data["weekday"] = data["date"].dt.weekday
# data["is_weekend"] = (data["weekday"] >= 5).astype(int)
```

```python
cols = data.columns.tolist()

# date 위치 찾기
date_idx = cols.index("date")

# year, month, day 제거 (어디에 있든)
for c in ["year", "month", "day"]:
    if c in cols:
        cols.remove(c)

# date 바로 오른쪽에 삽입
new_cols = (
    cols[:date_idx + 1]
    + ["year", "month", "day"]
    + cols[date_idx + 1:]
)

# 적용
data = data[new_cols]
```

```python
data.head()
```

```python
dong_mean = (
    data.groupby("dong_id")["electricity_kwh_day"]
    .mean()
)

data["dong_avg_power"] = data["dong_id"].map(dong_mean)
```

```python
data["y_delta"] = data["electricity_kwh_day"] - data["dong_avg_power"]
y = data["y_delta"]
data = data.rename(columns={"y_delta": "power_deviation"})

# 동 평균 대비 증감량
```

```python
features = [
    "dong_avg_power",
    "dong_id",
    "year",
    "month",
    "day",
    "temp_avg",
    "temp_min",
    "temp_max",
    "temp_range"
]

X = data[features]
y = data["power_deviation"]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    shuffle=False
)
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(
    n_estimators=40,
    max_depth=5,
    min_samples_leaf=5,
    random_state=21,
    n_jobs=-1
)
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print("RMSE:", rmse)
print("R2:", r2)
```
