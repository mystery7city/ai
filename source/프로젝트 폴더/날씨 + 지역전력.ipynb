```python
power1 = "data/결과물/power_region.csv.gz"

weather1 = "data/결과물/weather.csv.gz"
```

```python
import pandas as pd

power = pd.read_csv(
    power1,
    compression="gzip",
    encoding="utf-8"
)
```

```python
power_day = (
    power
    .groupby(["법정동코드", "법정동명", "date"], as_index=False)
    ["FDRCT_VLD_KWH"]
    .sum()
    .rename(columns={"FDRCT_VLD_KWH": "electricity_kwh_day"})
)
```
